### Pandas
pandas is a fast, powerful, flexible and easy to use open source data analysis and manipulation tool,
built on top of the Python programming language.

pandas extensively uses the functionality from numpy. Please familiarize yourself with numpy before delving into pandas. 

When working with tabular data, such as data stored in spreadsheets or databases, pandas is the right tool for you. pandas will help you to explore, clean, and process your data. In pandas, a data table is called a DataFrame.

Pandas offers two datastructures - DataFrame and Series. DataFrame for manipulation of tabular data and Series for manipulation of single column data.

#### Pandas Features

- pandas supports the integration with many file formats or data sources out of the box (csv, excel, sql, json, parquet,…). Importing data from each of these data sources is provided by function with the prefix read_*. Similarly, the to_* methods are used to store data.
- Selecting or filtering specific rows and/or columns? Filtering the data on a condition? Methods for slicing, selecting, and extracting the data you need are available in pandas.
- pandas provides plotting your data out of the box, using the power of Matplotlib. You can pick the plot type (scatter, bar, boxplot,…) corresponding to your data.
- You can derive new columns from existing columns - There is no need to loop over all rows of your data table to do calculations. Data manipulations on a column work elementwise. Adding a column to a DataFrame based on existing data in other columns is straightforward.
- Calculating summary statistics -Basic statistics (mean, median, min, max, counts…) are easily calculable. These or custom aggregations can be applied on the entire data set, a sliding window of the data, or grouped by categories. The latter is also known as the split-apply-combine approach.
- Changing structure of your dataframe - Change the structure of your data table in multiple ways. You can melt() your data table from wide to long/tidy form or pivot() from long to wide format. With aggregations built-in, a pivot table is created with a single command.
- Combining data from multiple dataframe -Multiple tables can be concatenated both column wise and row wise as database-like join/merge operations are provided to combine multiple tables of data.
- Great support for time series data - pandas has great support for time series and has an extensive set of tools for working with dates, times, and time-indexed data.
- Support for text data - pandas provides a wide range of functions to clean textual data and extract useful information from it.

```python
df_left.join(df_right, lsuffix='_caller', rsuffix='_other')
```

#### Data analysis example with pandas
#### using read_csv() to create a dataframe by reading data from a csv file
Dataset : The titanic.csv dataet from [here](https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv)

```python
import numpy as np
import pandas as pd

#use read_csv() to read the dataset into a dataframe called 'df', you can also read local files using read_csv()
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
```

```python
#Exploratory data analysis

print("\nShape of dataframe : ", df.shape)
print("\nColumn Names of dataframe : ", df.columns)
```

```python
# Describe statistical properties of each column numerical of dataframe
df.describe()
```

```python
# describe only the columns that are of type int64 or float64
df.describe(include=["int64", "float64"])
```

```python
# General information about dataframe - column name, number of non-null values in the column, datatype of the values in the column
df.info()
```

```python
#Statistical operations
# Mean age
print("Mean age: ", df['Age'].mean())

# Standard Deviation of age
print('Standard Deviation of age column: ', round(df['Age'].std(), 2))  #round() to round off std value to 2 decimal places
#Median PClass
print("Median PClass: ", df['Pclass'].mean())

#Total number of passengers who survived (Survived ==1)
print("Total number of passengers who survived (Survived ==1): ", df[df["Survived"] == 1]['Survived'].sum())

#Average age of passengers who survived
print("Average age of passengers who survived : ", round(df[df["Survived"] == 1]["Age"].mean(), 2))

#Maximum  age of male passengers who survived
#df[(df["Churn"] == 0) & (df["International plan"] == "No")]["Total intl minutes"].max()
print("Max age of male passengers who survived : ", 
      round( df[ (df["Survived"] == 1) & (df['Sex'] == 'male') ]["Age"].max(), 2))
```

```python
#The apply method can also be used to apply a function to each row. 

# Print all passengers whose name starts with 'B'
df[df["Name"].apply(lambda Name: Name[0] == "B")]
```

```python
#Print number of passengers whose name starts with 'B'
df[df["Name"].apply(lambda Name: Name[0] == "B")].shape[0]
```

```python
# here we display statistical properties of only age column for both groups of Survived.
df.groupby(["Survived"])['Age'].describe()
```

#### Pivot table and crosstab

pivot_table() helps create a pivot table from a dataframe. It need 3 parameters:
1. values – a list of columns for which statistical values are calculated
2. index –  a list of columns used to group data by,
3. aggfunc – functions which are applied to the groups of data created from columns in values parameter.


```python
# pivot table
# Use columns Age and Parch , group rows in these two columns by values of Pclass, apply max() to the values in each group
# So we see in the pivot table, for maximum values of Age and Parch for each unique value of Pclass column.

df.pivot_table(
    ["Age", "Parch"],
    ["Pclass"],
    aggfunc= "max"
)
```

```python
#Cross Table - helps us see the distribution of data in two or more column - it is like a contingency table
# here we see the number of data observations tabulated with respect to 2 unique values of Survived (along rows) and 3 unique values of Pclass(along columns) in dataset. 
pd.crosstab(df["Survived"], df["Pclass"])
```

```python
# Same as above but data is normalized
pd.crosstab(df["Survived"], df["Pclass"], normalize=True)
```

```python
# Changing datatype of some column
df["Pclass"] = df["Pclass"].astype("float64")
```

```python
# Find the number of unique values in a particular column
df["Pclass"].value_counts()
```

```python
#Sorting a dataframe : pandas dataframe can be sorted by values or by index.
#A DataFrame can be sorted by the values in one or more columns using sort_values() function
#Argument ascending=False sorts in descending order
#Here we sort the dataframe by value of Age column in descending order and print first five rows

df.sort_values(by="Age", ascending=False).head()
```

```python
# Here we sort by two columns  - PassengerId and Age both in ascending order
df.sort_values(by=["PassengerId", "Age"], ascending=[True, True]).head()
```

#### Changing contents of Dataframe


```python
# Adding a new column
# Here we create a new column "Sex_numerical" which contains a 1 for corresponding value 'male' in column 'Sex'
# and a value 2 for corresponding value 'female' in column 'Sex'
# we create a dictionary that maps strings male and female to numbers 1 and 2 respectively, then we apply this dictionary 
# to Sex column and store corrosponding values in a series called 'sex_numerical'
# we use insert() function to add this series as last column of our dataframe, name of the column is "Sex_numerical"

sex_dict = {'male': 1, 'female':2}
sex_numerical = df['Sex'].map(sex_dict)
df.insert(loc=len(df.columns), column="Sex_numerical", value=sex_numerical)
df.head()
```

```python
## Delete one or more columns or rows
## use drop(), specify axis parameter to choose the axis over which deletion should take place
## let us delete the column 'Embarked', axis = 1 means to delete along column axis
## you can also specify column number or row number instead of labels
df.drop(['Embarked'], axis = 1)
```
